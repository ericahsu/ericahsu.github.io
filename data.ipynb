```python
#Use tabula to convert the data tables from the county pdf into a csv

import tabula
import pandas as pd

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

file = "https://www.sandiegocounty.gov/content/dam/sdc/hhsa/programs/phs/Epidemiology/COVID-19%20Summary%20of%20Cases%20by%20Zip%20Code.pdf"

tabula.convert_into(file, "coviddata.csv", pages = "all", guess = True)

df = pd.read_csv('coviddata.csv')

df
```

```python
if len(df.columns) == 9:
    new_df = df.drop(df.columns[[2,5,7]], axis = 1) #drop empty columns
    #split the dataframe
    
    df1 = new_df.iloc[:, :3] 
    df2 = new_df.iloc[:, 3:]
    
    df2.rename(columns={'Rate per 100,000*.1':'Rate per 100,000*'}, inplace=True)

    df2.rename(columns={'Zip Code.1':'Zip Code'}, inplace=True)

    df2.rename(columns={'Count.1':'Count'}, inplace=True)
    
    df2.drop(df2.tail(2).index, inplace=True)

    final_df = pd.concat([df1, df2], axis = 0)

    final_df.reset_index(inplace = True)
    
    df3 = pd.read_csv('zipcodelatitudelongitude.csv')
    
    result = pd.concat([final_df, df3], axis = 1)
    
    result.rename(columns={'Rate per 100,000*':'Rate'}, inplace=True)
    
    result.to_csv ('finalcoviddata.csv')
    
    print(result)

else:
    df = pd.read_csv('coviddata.csv', skiprows=1)
    
    #drop extra line with header, bottom three rows with totals
    df.drop(df.index[54], inplace = True)

    df.rename(columns={'Rate per 100,000*':'Rate'}, inplace=True)

    df.drop(df.tail(2).index, inplace=True) # drop last 2 rows
    
    df.reset_index(inplace = True)

    df3 = pd.read_csv('zipcodelatitudelongitude.csv')

    result = pd.concat([df, df3], axis = 1)
        
    result.to_csv ('finalcoviddata.csv')

    print(result)

from datetime import datetime

date = datetime.today().strftime('%Y-%m-%d')

result.to_csv('coviddata'+date+'.csv')
```
